```python
df["Daily Return"] = df["Daily Return"].fillna(df["Daily Return"].mean())
df["ME50"] = df["ME50"].fillna(df["ME50"].mean())
df["MA200"] = df["MA200"].fillna(df["MA200"].mean())
df["Volatility"] = df["Volatility"].fillna(df["Volatility"].mean())
```

```python
df.isna().sum()
```

```python
df
```

```python
df["Profit/Loss"] = df["Daily Return"].apply(lambda x: "Profit" if x > 0 else "Loss")
```

```python
df["Profit/Loss"].value_counts()
```

```python
df["Profit/Loss"].value_counts().plot(kind="bar", figsize=(6,4))

plt.title("Profit vs Loss Days")
plt.xlabel("Result")
plt.ylabel("Number of Days")

plt.show()
```

```python
start_price = df["Close"].iloc[0]
end_price = df["Close"].iloc[-1]

years = (df.index[-1] - df.index[0]).days / 365

cagr = ((end_price / start_price) ** (1/years) - 1) * 100

print("CAGR:", round(cagr,2), "%")
```

```python
risk_free_rate = 0.02

avg_return = df['Daily Return'].mean()
Volatility = df['Daily Return'].std()

sharpe_ratio = (avg_return - risk_free_rate /252) / Volatility
print("sharpe ratio: ", round(sharpe_ratio,2))
```

```python
df["Peak"] = df["Close"].cummax()

df["Drawdown"] = (df["Close"] - df["Peak"]) / df["Peak"]

print("Maximum Drawdown:", df["Drawdown"].min()*100)
```

```python
plt.figure(figsize=(12,6))
plt.plot(df["Drawdown"])
plt.title("Market Drawdown (Crash Periods)")
plt.show()
```

```python
df["Market Type"] = df["Daily Return"].apply(
    lambda x: "Bull" if x > 0 else "Bear"
)

df["Market Type"].value_counts()
```

```python
forecast_days = 30

last_ma = df["ME50"].iloc[-1]

future_prices = [last_ma] * forecast_days
```

```python
plt.figure(figsize=(12,6))

plt.plot(df["Close"], label="Actual Price")

future_dates = pd.date_range(
    start=df.index[-1],
    periods=forecast_days+1,
    freq="D"
)[1:]

plt.plot(future_dates, future_prices, label="Forecast")

plt.legend()
plt.title("30-Day Price Forecast")
plt.show()
```
